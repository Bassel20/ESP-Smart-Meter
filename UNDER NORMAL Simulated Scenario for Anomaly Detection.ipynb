```python
df_train = pd.read_csv('traindata.csv', index_col="Datum Zeit")

df_test = pd.read_csv('testdata.csv', index_col="Datum Zeit")
```

```python
del df_train["Unnamed: 0"]
del df_test["Unnamed: 0"]
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = MinMaxScaler()
data_scaled = scaler.fit(df_train)
train_data_scaled = data_scaled.transform(df_train)
test_data_scaled = data_scaled.transform(df_test)
```

```python
from tensorflow.keras.models import Model

class AutoEncoder(Model):
  def __init__(self):
    super(AutoEncoder, self).__init__()
    self.encoder = tf.keras.Sequential([
                  tf.keras.layers.Dense(64, activation="relu"),
                  tf.keras.layers.Dense(32, activation="relu"),
                  tf.keras.layers.Dense(16, activation="relu"),
                  tf.keras.layers.Dense(8, activation="relu")
              ])
    self.decoder = tf.keras.Sequential([
                  tf.keras.layers.Dense(16, activation="relu"),
                  tf.keras.layers.Dense(32, activation="relu"),
                  tf.keras.layers.Dense(64, activation="relu"),
                  tf.keras.layers.Dense(140, activation="sigmoid")
              ])
  def call(self, x):
    encoded = self.encoder(x)
    decoded = self.decoder(encoded)
    return decoded
```

```python
import tensorflow as tf

model = AutoEncoder()
early_stopping = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=2, mode="min")
model.compile(optimizer='adam', loss="mae")
history = model.fit(train_data_scaled, train_data_scaled, epochs=50, batch_size=120,
                    validation_data=(test_data_scaled , test_data_scaled ),
                    shuffle=False,
                    callbacks=[early_stopping]
                    )
```

```python
encoder_out = model.encoder(train_data_scaled).numpy() #8 unit representation of data
decoder_out = model.decoder(encoder_out).numpy()
```

```python
plt.plot(train_data_scaled, 'b',)
plt.plot(decoder_out, 'r',)
plt.legend(["Original Normal Data", "Reconstucted Output"], loc ="upper right")
plt.title("Model performance on Normal data")


plt.show()
```

```python
encoder_out_a = model.encoder(test_data_scaled).numpy() #8 unit representation of data
decoder_out_a = model.decoder(encoder_out_a).numpy()
```

```python
plt.plot(test_data_scaled, 'b')
plt.plot(decoder_out_a, 'r')
plt.legend(["Original Anomalous Data", "Reconstucted Output"], loc ="upper right")
plt.title("Model performance on Anomaly Data")
plt.show()
```

```python
reconstruction = model.predict(train_data_scaled)
train_loss = tf.keras.losses.mae(reconstruction, train_data_scaled)
plt.hist(train_loss, bins=50)
```

```python
threshold = np.max(train_loss)
reconstruction_a = model.predict(test_data_scaled)
train_loss_a = tf.keras.losses.mae(reconstruction_a, test_data_scaled)
```
